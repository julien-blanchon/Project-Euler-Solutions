# Multiples of 3 and 5

## Problem 1

If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9. The sum of these multiples is 23.

Find the sum of all the multiples of 3 or 5 below 1000.

```python
solution = 0
max_value = 1000

for k in range(1,max_value//3+int(max_value%3)):
    print(3*k)
    solution += 3*k

for k in range(1,max_value//5+int(max_value%5)):
    print(5*k)
    solution += 5*k
    
for k in range (1,max_value//15+int(max_value%15)):
    print(15*k)
    solution -= 15*k

    
```

```python
def compute():
    ans = sum(x for x in range(1000) if (x % 3 == 0 or x % 5 == 0))
    return str(ans)
print(compute())
```
