```python
%%time

def prime_gen(max):
    a = 2
    previous_prime = []
    
    while a<max:
        a_can_be_prime = True
        while a_can_be_prime:
            
            for p in previous_prime:
                if a%p==0:
                    a_can_be_prime = False
                    break
                    
            if a_can_be_prime:
                previous_prime.append(a)
                a+=1
                yield previous_prime[-1]
            else:
                a+=1

solution = 1
max_val = 9000

for prime in prime_gen(max_val):
    i = 1
    print(prime)
    while prime**(i+1) <= max_val:
        i+=1
        print("\t" + str((prime**(i))))
    solution *= prime**(i)
    
solution
```
