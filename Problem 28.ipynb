```python
def primes_sieve(max_val):
    is_prime = [False]*2 + [True]*(max_val - 1) 
    for n in range(int(max_val**0.5 + 1.5)):
        if is_prime[n]:
            for i in range(n*n, max_val+1, n):
                is_prime[i] = False
    return [i for i, prime in enumerate(is_prime) if prime]
```
