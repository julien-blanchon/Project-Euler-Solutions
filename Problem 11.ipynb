```python
def product(l):
    result = 1
    for elt in l:
        result *= elt
    return result
```

```python
def grid_mask(l,x,y):
    x_max = len(l[0])-1
    y_max = len(l)-1
    if 0<=x<=x_max and 0<=y<=y_max:
        return l[y][x]
    else:
        return 0
```

```python
solution = []

for x in range(0,19):
    for y in range(0,19):
        for dx in (1,0,-1):
            if dx == 0:
                dy_range = (1,-1)
            else:
                dy_range = (1,0,-1)
            for dy in dy_range:
                                  
                solution.append([
                    product([grid_mask(grid,x,y),
                            grid_mask(grid,x+dx,y+dy),
                            grid_mask(grid,x+2*dx,y+2*dy),
                            grid_mask(grid,x+3*dx,y+3*dy)]),
                    [grid_mask(grid,x,y),
                    grid_mask(grid,x+dx,y+dy),
                    grid_mask(grid,x+2*dx,y+2*dy),
                    grid_mask(grid,x+3*dx,y+3*dy)],
                    [x,y]
                ])
                
                
```

```python
solution = sorted(solution, reverse=True)
print(solution[0])
```
